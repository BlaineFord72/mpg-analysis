```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mpg = sns.load_dataset('mpg')
```

Create a visual showing the relationship between horsepower and fuel economy.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='horsepower', y='mpg', data=mpg)
plt.title('Relationship between Horsepower and Fuel Economy')
```

Observations:
There is an inverse relationship between horsepower and fuel economy. As horsepower increases, fuel economy (mpg) generally decreases.
There is significant variation in fuel economy for cars with lower horsepower, while higher horsepower cars have more consistently low fuel economy.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mpg = sns.load_dataset('mpg')
```

```python
average_mpg_per_year = mpg.groupby('model_year')['mpg'].mean().reset_index()
```

Create a visual showing how fuel economy has changed over time

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='model_year', y='mpg', data=average_mpg_per_year, marker='o')
plt.title('Average Fuel Economy Over Time')
```

Observations:
The line plot shows a general upward trend in average fuel economy over the years, indicating overall improvements in fuel efficiency.
Dips in Specific Years- There are noticeable dips in fuel economy in the years 1973, 1975, and 1981. These could be due to various factors.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mpg = sns.load_dataset('mpg')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='weight', y='mpg', hue='origin', data=mpg)
plt.title('Weight vs. Fuel Economy Colored by Country of Origin')
```

Observations:
There is a negative correlation between the weight of the car and its fuel economy. Heavier cars tend to have lower mpg.
Cars from Japan tend to be lighter and have higher mpg compared to cars from the USA and Europe.
